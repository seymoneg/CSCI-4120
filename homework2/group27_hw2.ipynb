```python
# Starter code
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

from sklearn.datasets.samples_generator import make_blobs

# Instantiate synthetic dataset with 4 random clusters
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
```

```python
# TODO Determine the best k for k-means using the elbow method
# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12))

#Fits data to vizualizer and renders the image
visualizer.fit(X)
visualizer.show()
```

The best k for k-means using the elbow method is 4

```python
newModel = KMeans(4)
newModel.fit(X)
y_pred = newModel.predict(X)
```

```python
# TODO calculate accuracy for best K
```

```python
accuracy_score(y_true, y_pred)
```

```python
# TODO draw a confusion matrix
```

```python
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
mat = confusion_matrix(y_true, y_pred)
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('True Value')
plt.ylabel('Predicted Value')
```
